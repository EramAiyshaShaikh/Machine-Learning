Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS). **bold text**

Answer:
- Artificial Intelligence (AI)
  - The Big Idea: AI is the overarching concept of creating machines that can perform tasks requiring human intelligence, such as reasoning, problem-solving, and decision-making.
  - Scope: It's the broadest field among the four, aiming to mimic human-like intelligence.
- Machine Learning (ML)
  - A Method for AI: ML is a specific approach within AI that enables systems to learn and improve from data without being explicitly programmed.
  - How it Works: ML algorithms use statistical models and learn from experience to make predictions or decisions, improving over time with more data.
  - Relationship to AI: ML is a subset of AI, providing methods to achieve artificial intelligence.
- Deep Learning (DL)
  - A Specialized ML Technique: DL is a further specialization of Machine Learning that utilizes artificial neural networks (ANNs) with multiple layers.
  - How it Works: These deep neural networks are capable of recognizing complex patterns and learning hierarchical representations from vast amounts of data, mimicking the structure of the human brain.
  - Relationship to ML: DL is a subset of ML, and both are subsets of AI.
- Data Science (DS)
  - A Multidisciplinary Field: Data Science is a broad, interdisciplinary field focused on extracting actionable insights and knowledge from data.
  - Tools and Techniques: It employs a range of tools and techniques, including statistics, programming (like R and Python), data visualization, and often incorporates ML and DL algorithms to analyze data.
  - Human-Centric Goal: A key distinction is its human-centric goal of gaining understanding and insights that drive decision-making and strategic planning, often involving a human-in-the-loop.

Question 2: What are the types of machine learning? Describe each with one
real-world example. **bold text**

Answer:
- Types of machine learning:
  1. Supervised Learning
  - Description: This method involves training a model on a labeled dataset, where each data point has a known input and a corresponding output. The model learns to map these inputs to outputs and then uses this knowledge to predict outcomes for new, unseen data.
  - Real-world Example: Email spam detection. An algorithm is trained on a dataset of emails, each labeled as "spam" or "not spam". It learns patterns and features of spam emails to accurately classify new incoming emails as either spam or legitimate.
  2. Unsupervised Learning
  - Description: Unlike supervised learning, unsupervised learning deals with unlabeled data, meaning there are no predefined outputs for the inputs. The algorithm's goal is to find inherent patterns, structures, or groups within the data on its own.
  - Real-world Example: Customer segmentation. In marketing, unsupervised learning can group customers with similar purchasing behaviors or demographic information into different segments. This helps businesses tailor marketing strategies to specific customer groups.
  3. Semi-Supervised Learning
  - Description: This type of learning combines aspects of both supervised and unsupervised learning. It uses a small amount of labeled data along with a large amount of unlabeled data to improve its learning process and accuracy, often when labeling data is expensive or time-consuming.
  - Real-world Example: Personalized recommendation systems. Platforms can use a small amount of labeled data (e.g., user ratings) and combine it with a vast amount of unlabeled data about user interactions to provide more accurate and relevant recommendations.
  4. Reinforcement Learning
  - Description: Reinforcement learning involves an "agent" that learns to make decisions by interacting with an environment. The agent receives feedback in the form of rewards or penalties for its actions and learns to choose actions that maximize its cumulative reward over time.
  - Real-world Example: Game-playing AI. Algorithms like AlphaGo, which beat the world champion in the game of Go, learn through trial and error by playing millions of games. They receive rewards for making good moves and penalties for bad ones, eventually learning to play at a superhuman level to maximize their wins.

**Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

Answer:
- Underfitting (High Bias)
  - Definition: A model is too simple to represent the true relationship in the data.
  - Characteristics:
    High error on both training and test (unseen) data.
    Fails to learn the important patterns and trends in the data.
    Often caused by a lack of features, insufficient training, or too much regularization.
  - Example: A straight line trying to fit a dataset with a complex curved relationship.
- Overfitting (High Variance)
  - Definition: A model learns the training data too well, including its noise and random fluctuations.
  - Characteristics:
    Excellent performance on the training data but poor performance on new data.
    Excessively sensitive to the specific training data, leading to poor generalization.
    Often occurs with overly complex models.
  - Example: A high-degree polynomial that passes through every single data point, even the noisy ones.
- The Bias-Variance Tradeoff
  - The Goal: To build a model with low bias and low variance to achieve good generalization and accurate predictions on unseen data.
  - The Relationship:
  - Increasing model complexity: tends to decrease bias (better fit to training data) but increase variance (more sensitivity to training data fluctuations).
  - Decreasing model complexity: tends to increase bias (model too simple) but decrease variance (less sensitive to data noise).

Question 4: What are outliers in a dataset, and list three common techniques for
handling them. **bold text**

Answer:
- An outlier is an observation or data point that significantly deviates from the other data points in a dataset. Outliers can occur due to data entry errors, measurement mistakes, or genuine, but rare, natural variation. Properly handling outliers is crucial because they can skew statistical results, compromise the accuracy of machine learning models, and lead to misleading conclusions.
- Three common techniques for handling outliers
  1. Removal:
    For this technique, you simply delete the outlier data points from the dataset.
    - When to use: This method is best when you are certain the outlier is the result of a data entry error, a measurement mistake, or a faulty process. If outliers constitute a very small portion of your dataset (e.g., less than 2–3%), removal is often a simple and effective solution.
    - Limitations:
    It is not recommended for small datasets, as removing data can lead to a significant loss of valuable information.
    Removing legitimate, albeit extreme, observations can introduce bias and give an inaccurate representation of the overall data.
  2. Winsorization (Capping and Flooring):
    Winsorization limits the extreme values in a dataset by replacing the outliers with the nearest "non-outlier" value. For example, a value below the 5th percentile might be set to the 5th percentile value, while a value above the 95th percentile would be set to the 95th percentile value.
    - When to use: This technique is ideal when you want to minimize the influence of outliers without removing them entirely. It is a good choice if the extreme values are likely genuine, but you want to reduce their impact on your statistical analysis.
    - Limitations: Winsorization can introduce a slight bias by altering the data, but it is generally a safer approach than outright removal.
  3. Data transformation:
    This involves applying a mathematical function to the data to redistribute the values and reduce the influence of extreme observations. Common transformations include the logarithmic (log) or square root function.
    - When to use: Data transformation is especially useful when the dataset is skewed and includes extreme values. It can help normalize the data's distribution, making it more suitable for statistical models that assume a normal distribution.
    - Limitations: This method can complicate the interpretation of your results, as conclusions are drawn from the transformed data rather than the original values. It may not be effective for all types of distributions or outliers.

Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data. **bold text**

Answer:
- Process of Handling Missing Values
  1. Identify Missing Values: First, identify which columns in your dataset contain missing values and the extent of the missingness.
  2. Understand the Cause: Determine the potential reasons for the missing data, as this can influence the best imputation method. For example, if data is missing completely at random, simpler methods might suffice, but if it's related to other variables, more complex approaches are needed.
  3. Choose an Imputation Method: Select a strategy for filling the missing data. This choice depends on the data type (numerical or categorical) and the nature of the missingness.
  4. Implement the Imputation: Apply the chosen method to replace the missing values with the estimated values.
  5. Evaluate the Results: Check the impact of the imputation on the dataset's overall distribution and model performance.
- Numerical Imputation Technique: Mean/Median Imputation
  - Process: This method involves replacing missing numerical values with either the mean (average) or the median (middle value) of the column.
  - When to Use: Use the median when your data contains outliers, as the median is less sensitive to them than the mean. If there are no significant outliers, the mean is a simple and effective option.
- Categorical Imputation Technique: Mode Imputation
  - Process: For categorical data, missing values are replaced by the mode, which is the most frequently occurring category in the column.
  - When to Use: This is a straightforward way to handle missing categorical features, especially when the goal is to maintain the distribution of the existing categories.
    
    

```python
'''Question 6: Write a Python program that:
● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.
● Prints the class distribution.

Answer:
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, n_redundant=0, weights=[0.80])
display(X)
display(y)

df1 = pd.DataFrame(X,columns=['f1','f2'])
display(df1)

df2 = pd.DataFrame(y,columns=['target'])
display(df2)

final_df = pd.concat([df1, df2], axis=1)
display(final_df)

final_df['target'].value_counts()

plt.figure(figsize=(12,5))
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
plt.show()
```

```python
'''Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.

Answer:
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

no_samples = 1000
np.random.seed(1)
data = np.random.normal(loc=0, scale=1, size=no_samples)
df = pd.DataFrame(data, columns=['value'])
display(df)

missing_numbers = np.random.choice(no_samples, size=50, replace=False)
df.loc[missing_numbers, 'value'] = np.nan
display(df)

mean_value = df['value'].mean()
df['value_imputed'] = df['value'].fillna(mean_value)
display(df)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df['value'].dropna(), bins=30, edgecolor='black')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['value_imputed'], bins=30, edgecolor='black')
plt.title('Histogram After Imputation (Mean)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
'''Question 9: Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,
20] using sklearn.preprocessing.MinMaxScaler. Print the scaled array.

Answer:
'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

data = [2, 5, 10, 15, 20]
df = pd.DataFrame(data, columns=['value'])
display(df)

min_max = MinMaxScaler()
min_max_scaled_data = min_max.fit_transform(df)
scaled_data_df = pd.DataFrame(min_max_scaled_data, columns=['scaled_value'])
display(scaled_data_df)

final_df = pd.concat([df, scaled_data_df], axis=1)
display(final_df)
```

```python
'''Question 10: You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.

Answer:

Step 1: Handling Missing Data To address the missing ages, a common approach is to impute them.
The choice of imputation method depends on the data distribution.
The median is a robust choice as it is less affected by outliers than the mean.
Step 2: Addressing Outliers Outliers in the transaction amount can skew the model.
A standard method for identifying and handling them is the Interquartile Range (IQR) method.
Values that fall outside the range of \(Q1-1.5\times IQR\) and \(Q3+1.5\times IQR\) can be capped (winsorization) or removed.
Capping is often preferred to avoid losing data.
Step 3: Encoding Categorical Variables Categorical variables, such as payment method, need to be converted into a numerical format.
One-Hot Encoding is a suitable method for nominal categories like payment method, where no inherent order exists.
This creates a new binary column for each unique category.
Step 4: Addressing Class Imbalance A highly imbalanced target variable (fraud vs. non-fraud) can cause a model to be biased toward the majority class.
Techniques like SMOTE (Synthetic Minority Over-sampling Technique) can be used to generate synthetic examples of the minority class,
helping the model learn the patterns of fraudulent transactions more effectively.
'''

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler

data = {'age': [25, 30, np.nan, 45, 50, 60, 35, 70],
        'transaction_amount': [100, 150, 200, 500, 1000, 150, 20000, 80],
        'payment_method': ['card', 'paypal', 'card', 'crypto', 'card', 'paypal', 'card', 'crypto'],
        'is_fraud': [0, 0, 0, 0, 0, 0, 1, 0]}

df = pd.DataFrame(data)
display(df)

median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
display(df)

lower_bound = df['transaction_amount'].quantile(0.05)
upper_bound = df['transaction_amount'].quantile(0.95)
df['transaction_amount'] = np.clip(df['transaction_amount'], lower_bound, upper_bound).infer_objects(copy=False)
display(df)

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded = encoder.fit_transform(df[['payment_method']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['payment_method']))
df = pd.concat([df, encoded_df], axis=1)
df = df.drop('payment_method', axis=1) # Drop the original categorical column
display(df)


X = df.drop('is_fraud', axis=1)
y = df['is_fraud']
# Use RandomOverSampler for extreme imbalance
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)
smote_df = pd.concat([pd.DataFrame(X_res, columns=X.columns), pd.DataFrame(y_res, columns=['is_fraud'])], axis=1)

display(smote_df)
```
